## Знакомство с Gradio

**Gradio** - это библиотека Python с открытым исходным кодом, которая позволяет вам быстро создавать демонстрационные веб-приложения для вашей модели машинного обучения, а также создавать API и выполнять любые произвольные функции Python.

### Создание первого приложения на Gradio

Напишем первое приложение на Gradio, которое содержит 2 входных поля (текстовое и поле типа Слайдер) и одно выходное (текстовое) с результатам вычисления функции:

````python
import gradio as gr # импортируем библиотеку Gradio

# Наша вычислительная функция.
def greet(name, intensity):
    return "Меня зовут " + name + ". " + 'Мне ' + str(intensity) + ' лет!'

# создание интерфейса к функции
demo = gr.Interface(
    fn=greet,
    inputs=["text", "slider"],
    outputs=["text"],
)

# запуск приложения в режиме демонстрации
demo.launch(share=True)
````

Логика созданного приложения проста. Пользователь вводит свое имя в текстовое поле слева, перемещает ползунок в соответствии со своим возрастом и нажимает кнопку отправки. Справа появится результат выполнения функции.

Для этого просто введите gradio перед именем файла вместо python (достаточно ввести `gradio app.py` в своем терминале, где `app.py` - имя вашей программы.

В примере выше мы создали экземпляр класса `gr.Interface`. Класс `gr.Interface` предназначен для создания демонстраций моделей машинного обучения, которые принимают один или более входных данных и возвращают один или более выходных данных.

Класс `gr.Interface` имеет три основных аргумента:

* **fn** - функция для создания пользовательского интерфейса;
* **inputs** - входных данных (компонент Gradio), которые будут использоваться для ввода. Количество компонентов должно соответствовать количеству аргументов в вашей функции;
* **outputs** - выходные данные (компоненты Gradio), которые будут использоваться для вывода. Количество компонентов должно соответствовать количеству возвращаемых значений из вашей функции.

Аргумент **fn** очень гибкий - вы можете передать любую функцию Python, которую хотите обернуть пользовательским интерфейсом. В приведенном выше примере используется простая функцию, но эта функция может быть любой - от музыкального генератора до финансового калькулятора или функции прогнозирования предварительно обученной модели машинного обучения.

Входные и выходные параметры принимают один или несколько компонентов Gradio. Gradio включает в себя более 30 встроенных компонентов (таких как компоненты `gr.Textbox()`, `gr.Image()` и `gr.HTML()`), которые предназначены для приложений машинного обучения.

Если функция принимает более одного аргумента, как в случае, описанном выше, то необходимо передать список входных компонентов в **inputs**, где каждый входной компонент соответствует одному из аргументов функции, по порядку. То же самое справедливо, что если функция возвращает более одного значения, то необходимо передать список компонентов в **outputs**. Такая гибкость делает класс `Interface` очень эффективным способом создания демонстрационных приложений.

В качестве аргументов **inputs** и **outputs** вы можете передать название этих компонентов в виде строки ("textbox") или экземпляра класса (`gr.Textbox()`).


Перепишем пример выше, только вместо строк передадим экземпляры необходимых классов во входные и выходные функции:

````python
import gradio as gr # импортируем библиотеку Gradio

# Наша вычислительная функция.
def greet(name, intensity):
    return "Меня зовут " + name + ". " + 'Мне ' + str(intensity) + ' лет!'

# создание интерфейса к функции
demo = gr.Interface(
    fn=greet,
    inputs=[gr.Textbox(label="Имя", lines=1), gr.Slider(label="Возраст", value=18, minimum=1, maximum=100, step=1)],
    outputs=[gr.Textbox(label="Выходные данные", lines=1)],
)

# запуск приложения в режиме демонстрации
demo.launch(share=True)
````

Результат вывода:

image.png

В `gr.Textbox` и `gr.Slider` мы передали параметр `label` с подписью элемента формы (или его меткой).

Элемент `Textbox` имеет параметр `lines`, который определяет сколько строк в высоту будет занимать текстовое поле. Если будет больше строк использовано, чем данный параметр, то появится полоса прокрутки.

Для `Slider` мы определили параметры: `value` - начальное значение, `minimum` - минимальное значение, `maximum` - максимальное значение, `step` - шаг изменения значений при перетаскивании "ползунка" слайдера.

### Что еще умеет Gradio?

До сих пор мы обсуждали класс `Interface`, который является высокоуровневым классом, позволяющим быстро создавать демонстрации с помощью Gradio.

Но это не все, что умеет Gradio. Приведем более интересные возможности описанные в документации Gradio:

* **Чат-боты**. Gradio включает в себя еще один высокоуровневый класс `gr.ChatInterface`, который специально разработан для создания пользовательского интерфейса чат-бота. Аналогично `Interface`, вы вводите функцию, и Gradio создает полностью работающий пользовательский интерфейс чат-бота.

* **Настраиваемые блоки**. Gradio также предлагает низкоуровневый подход к разработке веб-приложений с более гибкими макетами и потоками данных с помощью класса `gr.Blocks`. `Blocks` позволяет выполнять такие действия, как управление отображением компонентов на странице, обработка сложных потоков данных (например, выходные данные могут служить входными данными для других функций) и обновление свойств / видимости компонентов на основе взаимодействия с пользователем. `gr.Blocks` позволяет создавать индивидуальные и сложные приложения.

* **Собственная экосистема**. На самом деле Gradio - это целая экосистема библиотек Python и JavaScript, которые позволяют создавать приложения для машинного обучения или выполнять к ним программные запросы не только на Python, но и на JavaScript. Т.е. если вы разместите свое приложение на [Hugging Face Space](https://huggingface.co/spaces), то сможете обращаться к нему по имени, как к функции, используя библиотеки:
  * `gradio_client` - программно запрашивает любое приложение Gradio на Python (см. [пример](https://www.gradio.app/guides/getting-started-with-the-python-client)).
  * `@gradio/client` - программно запрашивает любое приложение Gradio на JavaScript (см. [пример](https://www.gradio.app/guides/getting-started-with-the-js-client)).  
  * `@gradio/lite` - библиотека, которая использует [Pyodide](https://pyodide.org/en/stable/) для создания Gradio непосредственно в вашем браузере (см. [пример](https://www.gradio.app/guides/gradio-lite)).

### Чат-бот на Gradio

Чат-бот - это приложение для взаимодействия с большими языковыми моделями (LLM). С помощью Gradio вы можете легко создавать демонстрационные версии своей языковой модели и делиться ею со своими заказчиками, используя интуитивно понятный интерфейс чат-бота.

Для этих целей в Gradio имеется класс `gr.ChatInterface()`. Это высокоуровневая абстракция, позволяющая быстро создать пользовательский интерфейс чат-бота с помощью одной строчки кода:
```python
gr.ChatInterface(predict).launch()
```
Здесь `predict` - ваша функция предсказания.


Для описания функции предсказания, нам необходимо вспомнить урок [Telegram бот c ChatGPT на борту. Знакомство с  ChatGPT](https://colab.research.google.com/drive/1tqXt0XNstMb2TeJ8Za9S2xQA6ipjyMPk), где мы работали с API от OpenAI. Вам необходимо снова запросить у куратора секретный ключ для доступа к API OpenAI.

В примере ниже используется экспериментальный доступ к ChatGPT 4 Omni. Если он окажется недоступным, то замените `model='gpt-4o'` на `model='gpt-3.5-turbo'`.

````python
from openai import OpenAI # для работы с API OpenAI
import gradio as gr
import getpass # для работы с паролями

# Запрос ввода ключа от OpenAI
api_key = getpass.getpass("Введите OpenAI API Key:")

# Создание клиента к API OpenAI
client = OpenAI(api_key=api_key)

# Функция предсказания с двумя параметрами
#  message - текущее сообщение
#  history - история сообщений с чат-ботом
def predict(message, history):
    history_openai_format = [] # список сообщений в формате пригодном для OpenAI
    for human, assistant in history: # перебираем все сообщения
        # Указываем для каждого сообщения роль
        history_openai_format.append({"role": "user", "content": human }) # наши запросы
        history_openai_format.append({"role": "assistant", "content": assistant}) # ответы чат-бота
    # Наше последнее сообщение для чат-бота
    history_openai_format.append({"role": "user", "content": message})

    #  Формируем запрос к чат-боту со всей историей переписки
    response = client.chat.completions.create(
        model='gpt-4o',                   # используемая модель
        messages= history_openai_format,  # список форматированных сообщений с ролями
        temperature=1.0                   # точность ответов модели
                                              )

    return response.choices[0].message.content # возвращаем результат предсказания

# Запуск интерфейса чат-бота с функцией предсказания
gr.ChatInterface(predict).launch()
````

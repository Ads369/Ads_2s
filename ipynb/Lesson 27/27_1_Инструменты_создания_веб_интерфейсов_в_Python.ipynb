## Streamlit, Gradio, Plotly Dash, Panel - обзор инструментов

Модели машинного обучения сами по себе интересны только с научной точки зрения или развлечения. Как только модель будет завершена, ее, вероятно, придется развернуть, прежде чем она сможет принести какую-либо ценность. Кроме того, чрезвычайно полезна возможность развернуть предварительную модель или прототип для получения отзывов от других заинтересованных сторон.

Неслучайно все большую популярность среди специалистов в сфере данных набирают инструменты создания прототипов интерфейсов, позволяющие быстро и легко развертывать модели машинного обучения. Наиболее популярными считаются: Streamlit, Gradio, Plotly, Dash, Panel и Anvil.

Пройдемся кратко по плюсам и минусам каждой из них.

**Streamlit** — это [библиотека](https://streamlit.io/) позволяет развернуть приложение на Python в общедоступное веб-приложение. [Streamlit](https://streamlit.io/) объединил вокруг себя сообщество специалистов по обработке данных и машинному обучению, позволяет создавать информационные панели и интерактивные модели машинного обучения. Простота Streamlit сделала его столь популярным! Использование библиотеки Streamlit является быстрым и интуитивно понятным,  вы можете создавать приложения для обработки данных, не имея опыта веб-разработки.

image.png

**Рис. 1. Пример интерфейса Streamlit**

Streamlit можно запускать как библиотеку на своем веб-сервере, так и запускать в специализированном облаке [Streamlit Community Cloud](https://docs.streamlit.io/streamlit-cloud/get-started/deploy-an-app). Streamlit Community Cloud запускает приложения непосредственно из вашего репозитория GitHub, поэтому код вашего приложения и зависимости должны находиться на GitHub, прежде чем вы попытаетесь развернуть свое приложение.

Однако у Streamlit есть и свои недостатки. Его макет и стиль приложений довольно строгии, поэтому, если вы не хотите заниматься веб-версткой с применением HTML, JS и CSS, то все ваши приложения будут выглядеть практически одинаково. У Streamlit также есть проблемы со скоростью, поскольку он повторно запускает приложение Python, после каждого взаимодействия с интерфейсом.

**Gradio** — это [библиотека](https://gradio.app/) Python для превращения наборов данных и моделей машинного обучения в интерактивные веб-интерфейсы. Основное отличие от Streamlit заключается в том, что Gradio больше предназначен для демонстраций машинного обучения, тогда как Streamlit предназначен для создания информационных панелей данных.

image.png

**Рис. 2. Пример интерфейса Gradio**

Его самое большое преимущество заключается в том, что oн теперь является частью [Hugging Face](https://huggingface.co/), а интеграция Hugging Face обеспечивает легкую и удобную загрузку моделей и наборов данных ML.

Еще одно преимущество Gradio заключается в том, что с его помощью легко делиться своими приложениями. Вы можете быстро создать ссылку для отправки другим пользователям, но срок действия этой ссылки истекает через 72 часа, и весь код выполняется на вашем компьютере. Чтобы поделиться своим приложением, вам нужно разместить его где-нибудь еще, например [Hugging Face Spaces](https://huggingface.co/spaces).

Среди минусов можно отметить его устаревший внешний вид, его, конечно же, можно стилизовать, но для этого придется глубоко погрузиться в язык каскадных стилей CSS.

Другой минус менее очевиден для разработки, но более существенен с маркетинговой точки зрения. Так как Gradio интегрирован в Hugging Face, то приложения, размещенные в Spaces, размещаются на веб-сайте Hugging Face по жестким правилам, что затрудняет брендирование вашего приложения.

Если сравнивать Gradio и Streamlit, то важно уяснить один аспект: Gradio не является автономным приложением, поэтому его необходимо встраивать в другую веб-страницу, в отличие от приложений Streamlit, которые хорошо работают как автономные приложения.

**Plotly** — это [библиотека](https://dash.plotly.com/) построения графиков, основанная на платформе **Dash** с открытым исходным кодом, которая умеет работать не только с Python, но и с такими языками как R или Julia.

image.png

**Рис. 3. Пример интерфейса Plotly**

Plotly быстрее, чем Streamlit. Ускорение достигается за счет того, что Dash запускает только те функции, которые вызываются во время взаимодействия с приложением. При использовании же Streamlit все приложение перезапускается при каждом взаимодействии.

Plotly имеет глубокую интеграцию в блокнотах Jupyter. Если вы предпочитаете работать в блокноте Jupyter, то библиотека jupyter-dash позволяет запускать информационные панели прямо в блокноте Jupyter.

В отличии от Streamlit и Gradio, Plotly имеет более гибкий дизайн. Даже не глубокие знания в HTML и CSS, позволят вам стилизовать свои приложения Dash так, как вы хотите.

Но все же Dash немного сложнее, чем Streamlit, как минимум вам нужно будет научиться использовать «обратные вызовы» и немного подтянуть HTML, поэтому "порог входа" здесь выше.

**Panel** — это [библиотека](https://panel.holoviz.org/) Python с открытым исходным кодом для создания веб-приложений интерактивной информационной панели. Panel чрезвычайно гибка и позволяет использовать любую библиотеку графиков, которая вам нравится. Как и в Gradio вы можете использовать Panel в блокнотах Jupyter. Панели мониторинга Panel также можно развертывать как отдельные веб-приложения, но, как и в случае с Plotly Dash, вам потребуется настроить сервер для его развертывания самостоятельно.

image.png

**Рис. 4. Пример интерфейса Panel**





В Panel есть несколько готовых визуальных шаблонов для быстрого старта, стилизация настраивается проще чем Streamlit, да и работает гораздо быстрее.

Главный минус Panel, что это небольшое сообщество вокруг библиотеки, поэтому находить информацию сложнее, да и сама библиотека довольно сложная, нет готового варианта облачного хостинга. Panel позволяет легко запускать и размещать ваши приложения локально, но если вы хотите развернуть их в облаке, то сможете сделать это самостоятельно.

**Что выбрать?**

* **Streamlit**, если вы хотите быстро запустить панель мониторинга и иметь возможность добавлять множество компонентов и элементов управления. Кроме того, Streamlit позволяет создавать веб-интерфейс или панель мониторинга намного быстрее, чем конкуренты. Быстр в разработке, но медленный в работе!
* **Gradio**, если хотите использовать всю мощь пакетов HuggingFace и для вас не важны автономность и внешний вид. Хотя, как говорится: "На вкус и цвет все фломастеры разные". Gradio специально создавался с учетом моделей машинного обучения. Поэтому, если вы хотите создать веб-интерфейс специально для созданной вами модели машинного обучения, вам подойдет простой синтаксис и настройка Gradio.
* **Plotly**, если вы создаете интерфейс, где скорость играет важную роль, хотите иметь возможность разрабатывать на разных языках (а не только python) и с поддержкой блокнота Jupyter. Также ваш уровень HTML и CSS выше начального и вы готовы потратить чуть больше времени на обучение. Платформа Dash чаще всего используется для создания информационной панели для более крупных компаний, поскольку она в основном и разрабатывалась для корпоративных решений.
* **Panel**, если хотите использовать произвольную библиотеку построения графиков и визуализации, а не ту которую использует данный инструмент. Для вас не важна интеграция с готовым облачным хостингом, вы разрабатываете в блокноте Jupyter и при необходимости сами сможете настроить хостинг.

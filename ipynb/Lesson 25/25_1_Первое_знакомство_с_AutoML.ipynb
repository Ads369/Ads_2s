##Как работает AutoML?

Как правило, рабочий процесс для обучения моделей машинного обучения выглядит следующим образом:

1. Определение проблемы
2. Сбор данных
3. Предварительная обработка данных
4. Обучение модели
5. Оценка модели

Предварительная обработка, обучение и оценка — это экспериментальный и итеративный (несколько последовательных шагов с изменением параметров) процесс, который требует нескольких испытаний, пока вы не достигнете удовлетворительных результатов. Так как эти задачи, как правило, повторяются, AutoML может помочь автоматизировать эти шаги. Помимо автоматизации, методы оптимизации используются во время процесса обучения и оценки для поиска и выбора алгоритмов и гиперпараметров.

**Описание набора данных**

Данный датасет состоит из двух файлов:
* **train.csv** - обучающий набор, содержит колонку `isFraud` - признак мошеннической операции (бинарный параметр 0 или 1);
* **test.csv** - тестовый набор, не содержит колонку `isFraud`, ее и требуется предсказать.

Данные в наборе можно разбить на 2 части `identity` (идентификационную информацию) и `transaction` (информацию о транзакции). Не все транзакции содержат соответствующую ей идентификационную информацию в датасете.

Информация о транзакциях представлена колонками:

* **TransactionDT**  - время, отсчитываемое от заданной контрольной точки, некоторой даты и времени (не фактическая временная метка);

* **TransactionAMT** - сумма платежа по транзакции в долларах США (USD);

* **ProductCD** - код продукта для каждой транзакции;

* **card1** - **card6** - информация о платежной карте, такая как тип карты, категория карты, банк-эмитент, страна и т.д.

* **addr** - адрес;

* **dist** - расстояние, удаленность;

* **P_emaildomain** и **R_emaildomain** - домен электронной почты покупателя и получателя;

* **C1**-**C14** - расчетные данные, например, количество найденных адресов, связанных с платежной картой, и т.д. Фактический смысл скрыт как банковская тайна.

* **D1**-**D15** - расчетные временные интервалы, например, дни между предыдущими транзакциями и т.д. Фактический смысл скрыт как банковская тайна.

* **M1**-**M9** - расчетные величины соответствий, например, имен на карточке и адреса и т.д. Фактический смысл скрыт как банковская тайна.

* **Vxxx** - множество функций, включая ранжирование, подсчет и другие взаимосвязи между объектами. Расчетные величины. Фактический смысл скрыт как банковская тайна.

Также датасет содержит и идентификационные данные – информация о сетевом подключении (IP, провайдер, прокси-сервер и т.д.) и цифровая подпись (пользовательский адрес/браузер/операционная система/версия и т.д.), связанные с транзакциями. Они собираются системой защиты от мошенничества банка, а также сторонними организациями-партнерами банка по цифровой безопасности. Эти данные носят фрагментированный характер (что-то есть, а чего-то нет).


**ВАЖНО**. Вы учитесь работать с набором данных "вслепую". Названия полей замаскированы, попарный словарь для восстановления данных не предоставляется в целях защиты конфиденциальности. Подробное и внятное название поле, их описание отсутствует. То есть приходится работать с тем, что есть - все как в "реальной" практике. Большего описания вам банк не даст, так как боится случайно "раскрыть" банковскую тайну, поэтому лучшая позиция для банка: "это все, что у нас есть, мы вам все материалы передали!"

Первой библиотекой, которая нам поможет разобраться в наших "непонятных" и "сильно засекреченных" данных, будет библиотека `H2O`. Большой плюс библиотеки в том, что она самостоятельно проводит предобработку данных и нам нет необходимости проводить нормализацию данных.

### Библиотека H20

[H2O AutoML](https://docs.h2o.ai/h2o/latest-stable/h2o-docs/automl.html) является открытой платформой для анализа данных и машинного обучения, обеспечивает широкий спектр алгоритмов машинного обучения, включая глубокое обучение, градиентный бустинг, случайный лес и многие другие.

AutoML в H2O работает по следующей схеме:

1. **Загрузка данных**: В первую очередь вам нужно загрузить свой набор данных в H2O.

2. **Запуск AutoML**: Далее вы можете запустить процесс AutoML, указав, какие столбцы данных использовать в качестве входных признаков, а какой - в качестве целевого. Вы также можете указать, сколько времени вы хотите, чтобы AutoML работал.

3. **Подбор моделей**: AutoML автоматически обучает и валидирует множество моделей, используя ваш набор данных. Он использует различные алгоритмы и методы для построения и оптимизации моделей, включая предобработку данных, отбор функций, настройку гиперпараметров.

4. **Ранжирование моделей**: По завершении процесса AutoML модели ранжируются в соответствии с их производительностью на основе выбранной метрики.

5. **Использование модели**: Вы можете выбрать лучшую модель для дальнейшего использования, а также получить ее параметры и информацию о производительности.

```python
import numpy as np # знакомая нам библиотека, как ее часто называют -  линейной алгебры
import pandas as pd # для работы с наборами данных (датафреймы)
import os # работа с файловой системой
```

```python
import h2o                            # загружаем библиотеку H2O
print(h2o.__version__)                # выводим версию библиотеки
from h2o.automl import H2OAutoML      # выводим версию библиотеки


h2o.init(max_mem_size='16G')          # запускаем локально кластер H2O, выделяем 16Gb оперативы под кластер
```

#### Загрузка датасета

```python
%time # магическая функция для оценки времени выполнения

# загружаем датасет в датафрейм (датафрейм H2O отличается от Pandas датафрейм)
train = h2o.import_file("./dataset/train.csv")
test = h2o.import_file("./dataset/test.csv")
```

Выведем первые 10 строк датафрейма:

```python
train.head()
```

```python
# Определение предикторов (параметры по которым предсказываем) и целевой переменной
x = train.columns     # Получаем список всех колонок в данных обучения
y = 'isFraud'         # Указываем имя целевой переменной
x.remove(y)           # Удаляем из списка предикторов имя целевой переменной

# Для бинарной классификации целевая переменная должна быть фактором (категорией)
train[y] = train[y].asfactor() # Преобразуем целевую переменную в фактор
```

#### Запуск AutoML для подбора моделей и их обучения

```python
# Запуск AutoML для 15 базовых моделей (max_models)
# Не более часа на подбор моделей (max_runtime_secs)
# seed - начальное значение для генератора случайных чисел (для воспроизводимости результатов)
aml = H2OAutoML(max_models=15, seed=42, max_runtime_secs=3600)

 # Обучаем модели на заданном обучающем датасете (training_frame) с указанными предикторами (x) и целевой переменной (y)
aml.train(x=x, y=y, training_frame=train)
```

В H2O AutoML метрика, которую нужно оптимизировать, выбирается автоматически на основе типа задачи.

Если задача является задачей классификации, как в нашем случае, H2O AutoML будет автоматически оптимизировать метрику `logloss`. Для задач регрессии используется `deviance`.

**ВАЖНО**. H2O AutoML всегда стремится максимизировать метрику, поэтому для метрик, где меньшие значения лучше (например, `logloss` или `MSE`), H2O использует их отрицательные значения.

Если вы хотите использовать конкретную метрику для оптимизации, то необходимо задать параметр `stopping_metric` (в методе `aml.train()`) при создании экземпляра H2OAutoML. По умолчанию **AUTO** выбирается `logloss` для классификации и `deviance` для регрессии. Также доступные значения: `deviance` (среднее остаточное отклонение), `logloss`, `MSE`, `RMSE`, `MAE`, `RMSLE`, `AUC` (площадь под ROC-кривой), `AUCPR` (площадь под кривой точность-полнота), `lift_top_group`, `misclassification`, `mean_per_class_error`.

#### Оценка моделей

 В результате обучения мы получили ни одну модель, а объект AutoML (переменная `aml`), который содержит до 15 обученных моделей (сколько успеет за час обучить, скорее всего 2-3 модели), основанных на различных алгоритмах машинного обучения.

Переменная `aml` также содержит "доску лидеров" (`leaderboard`) - таблицу, которая ранжирует все модели, обученные AutoML, на основе выбранной метрики.

Самая сильная модель (с лучшим значением метрики) становится "лидером" и может быть использована для предсказания новых данных. Вы можете получить эту модель, используя `aml.leader`.

```python
# Выводим доску лидеров
lb = aml.leaderboard
lb.head(rows=lb.nrows)  # Выводим все строки лидеров
```

В рамках урока мы не будем детально вникать в метрики, основное достоинство AutoML - это автоматический выбор наиболее оптимальной метрики и оптимизации параметров с учетом выбора.

На "доске лидеров" представлены данные на проверочной выборке (H2O самостоятельно поделила выборки и провела оценку на проверочной выборке).

Для задачи классификации по умолчанию оптимизировалась метрика `LogLoss` (чем меньше, тем лучше), а точность мы можем оценить по метрике на валидационной выборке. Для лучшей модели получена метрика `AUC` - 0.956. Ее чаще всего называют ROC-AUC. И это самая популярная метрика для бинарной классификации в классическом машинном обучении. И хорошо подходит для несбалансированной выборки. Чем она ближе к 1, тем лучше. Для любознательных подготовлен отдельный коллаб по этой метрике ([ROC-AUC](https://colab.research.google.com/drive/1G6PfW_36EOBgjVTy3ako29uQugu3YTeY)).

```python
# Лучшая модель (лидер)
aml.leader
```

Удаляем из памяти обучающую выборку, экономим ресурсы.

```python
h2o.remove(train)
```

Сохранить лучшую модель для дальнейшего использования в бинарном формате:

```python
best_model=aml.leader
model_path = h2o.save_model(model=best_model, path="./model_fraud", force=True)
```

Загружаем модель в H2O:

```python
loaded_model = h2o.load_model(path=model_path)
```

Также можем сохранить данные в формате MOJO, часто используется для разворачивания модели на конечных устройствах или веб-сервере.

```python
# сохранить лучшую модель в формате MOJO можно так:
best_model.download_mojo(path = "./model_fraud")
```

Далее мы проведем предсказание на тестовой выборке, что транзакция является мошеннической:

```python
preds = loaded_model.predict(test)
```

```python
preds.head()
```

И добавим в первую колонку тестовой выборки предсказанный результат:

```python
test[y] = preds['predict']
```

## Основная концепция Streamlit

Приложение Streamlit — это ваш холст, на котором вы будете рисовать диаграммы, текст, виджеты, таблицы и многое другое.

Что будет нарисовано в приложении, зависит от вас. Например, `st.text()` записывает необработанный текст в ваше приложение, `st.line_chart()` рисует линейную диаграмму.



### Отображение и стилизация данных

Существует несколько способов отображения данных (таблиц, массивов, фреймов данных) в приложениях Streamlit.

Знакомство с Streamlit всегда начинается с "магической функции" `st.write()`, с помощью которых можно писать практически все, что угодно — от текста до таблиц. Почему они называются магическими? А потому, что вся магия в том, что вы их можете даже не писать! Streamlit сам их за вас выполнит.





````python
%%writefile main.py
import streamlit as st
import pandas as pd
df = pd.DataFrame({
  'first column': [1, 2, 3, 4],
  'second column': [10, 20, 30, 40]
})

df
````

выполнилась так, как-будто мы указали:

Поэтому метод `st.write` называют магическим.

Каждый раз, когда Streamlit видит переменную или буквальное значение в отдельной строке, он автоматически оборачивает его в `st.write`.

Вы можете передать практически все в `st.write`: текст, данные, модель Keras, функции, диаграммы Matplotlib, фигуры Plotly и многое другое. Независимо от данных, Streamlit разберется во всем и отобразит все правильно.

Существуют и другие функции, специфичные для данных, например, `st.dataframe()` и `st.table()`, которые вы также можете использовать для отображения данных.

Давайте разберемся, когда есть смысл выбрать альтернативу магическому методу `st.write()`:
1. Вам нужна статическая таблица - используйте `st.table()`.
2. `st.write()` и `st.dataframe()` рисуют интерактивные таблицы, они кликабельные, их можно редактировать.
3. Используйте `st.dataframe()`, если вы хотите передать дополнительные аргументы для настройки его поведения.

Например, `st.dataframe()` умеет работать с форматированием  [Styler](https://pandas.pydata.org/docs/user_guide/style.html) объекта Pandas.

Запустим демонстрацию всех трех методов (не забываем прервать выполнение ячейки выше).

````python
%%writefile main.py
import streamlit as st
import pandas as pd
df = pd.DataFrame({
  'first column': [1, 2, 3, 4],
  'second column': [10, 20, 30, 40],
  'new column': [7, 9, 10, 20]
})

st.header("Демонстрация методов отображения данных:") # Вывод заголовка
st.subheader("Пример st.write(df):")                  # Вывод подзаголовок
st.write(df)

st.subheader("Пример st.dataframe(df):")
st.dataframe(df.style.highlight_max(axis=0))          # Подсветим строку с максимальными значениями (см. Styler Pandas)

st.subheader("Пример st.table(df):")
st.table(df)
````

Для вывода различной текстовой информации есть и специальные методы:

* `st.title()` - название приложения;
* `st.header()` - отображение заголовка раздела;
* `st.subheader()` - отображение заголовка подраздела (заголовок 2-го уровня);

### Рисование графиков и карт

Streamlit поддерживает несколько популярных библиотек диаграмм данных, таких как Matplotlib, [Altair](https://altair-viz.github.io/) и другие.

#### Простые диаграммы

Вы можете легко добавлять диаграммы в свои приложения Streamlit:

* `st.line_chart()` - линейная диаграмма;
* `st.area_chart()` - диаграмма с областями;
* `st.bar_chart()`  - гистограмма.

Cгенерируем случайную выборку с помощью Numpy и построим все три диаграммы.

С помощью метода `st.columns(3)` создадим 3 колонки и заполним их диаграммами. Иначе диаграммы отобразятся вертикально: одна под другой.

````python
%%writefile main.py
import streamlit as st
import numpy as np
import pandas as pd

chart_data = pd.DataFrame(
     np.random.randn(20, 3),
     columns=['a', 'b', 'c'])


col1, col2, col3 = st.columns(3)

col1.line_chart(chart_data) # линейная диаграмма;
col2.area_chart(chart_data) # диаграмма с областями;
col3.bar_chart(chart_data)  # гистограмма;
````

### Диаграммы matplotlib

`st.pyplot()` - позволяет строить диаграммы с помощью библиотеки **matplotlib**.

````python
%%writefile main.py
import streamlit as st
import matplotlib.pyplot as plt

data = {'apple': 10, 'orange': 15, 'lemon': 5, 'lime': 20}
names = list(data.keys())
values = list(data.values())

fig, axs = plt.subplots(1, 3, figsize=(9, 3), sharey=True)
axs[0].bar(names, values)
axs[1].scatter(names, values)
axs[2].plot(names, values)
fig.suptitle('Фрукты')

st.pyplot(fig)
````

**Элемент управления "выбор"**

Для выбора значения из списка используется `st.selectbox()` или `st.radio()`. Первый параметр - это метка, сообщение пользователю, что делает данный элемент управления. Второй параметр - значения для выбора, вы можете задать их списком, кортежем, либо передать их через массив или столбец фрейма данных.

Давайте воспользуемся `df` фреймом данных, который мы создали ранее и будем выбирать значения для `st.selectbox()` из `df['first column']`, а для `st.radio()` из `df['second column']`

````python
%%writefile main.py
import streamlit as st
import pandas as pd

df = pd.DataFrame({
  'first column': [1, 2, 3, 4],
  'second column': [10, 20, 30, 40],
  'new column': [7, 9, 10, 20]
    })

option = st.selectbox(
    'Какое значение вам нравится больше всего?',
     df['first column'])

'Ваш выбор: ', option  # Еще одна магия, подразумевается st.write

chosen = st.radio(
        'Какое значение вам нравится больше всего?',
        df['second column'])

'Ваш выбор: ', chosen  # Еще одна магия, подразумевается st.write
````

## Макет

Streamlit позволяет легко организовать виджеты на боковой панели с помощью `st.sidebar`. Каждый передаваемый элемент `st.sidebar` закрепляется слева, что позволяет пользователям сосредоточиться на содержимом вашего приложения, сохраняя при этом доступ к элементам управления пользовательского интерфейса.

Например, если вы хотите добавить поле выбора и ползунок на боковую панель, используйте `st.sidebar.slider()` и `st.sidebar().selectbox()` вместо `st.slider()` и `st.selectbox()`:

````python
%%writefile main.py
import streamlit as st

# Добавляем selectbox для sidebar:
add_selectbox = st.sidebar.selectbox(
    'Как с вами можно связаться?',
    ('Email', 'Home phone', 'Mobile phone')
)

# Добавляем slider для sidebar:
add_slider = st.sidebar.slider(
    'Выберите значение диапазона',
    0.0, 100.0, (25.0, 75.0)
)
````

Помимо боковой панели, Streamlit предлагает несколько других способов управления макетом вашего приложения.

`st.columns()` позволяет размещать виджеты рядом, а `st.expander()` экономить место, скрывая большой контент.

````python
%%writefile main.py
import streamlit as st

left_column, right_column = st.columns(2) # создаем колонки

if left_column.button('Нажми меня!'): # так можно добавить кнопку
    # Действие при нажатии кнопки
    st.write("Кнопка нажата!")


# Чтобы лишний раз не писать right_column.radio, можно вызывать нужные элементы в блоке with:
with right_column:
    # Элемент радио-кнопка, выбор одного из значения
    chosen = st.radio(
        'Какие-то значение',
        ("1", "2", "3", "4"))
    st.write(f"Какое-то значение {chosen} выбрано!")
````

## Прогресс выполнения

Добавляя в приложение длительные вычисления, вы можете использовать `st.progress()` для отображения статуса в реальном времени.

Для имитации долгих вычислений используем метод `sleep` библиотеки `time`:




````python
%%writefile main.py
import streamlit as st
import time

'Запуск долгого вычисления...' # магическая функция

# Добавляем счетчик итераций
latest_iteration = st.empty()

# Добавляем элемент прогресса, начальное значение 0'
bar = st.progress(0)

for i in range(100):
  # Обновляем элемент прогресса каждую итерацию.
  latest_iteration.text(f'Итерация {i+1}') # новое значение счетчика итераций
  bar.progress(i + 1) # новое значение прогресса
  time.sleep(0.1) #

'...миссия выполнена!' # и снова магическая функция
````
